```python
dados_requisicao, dados_item, dados_prestador, dados_beneficiario, 
dados_requisicao_item.DS_STATUS_ITEM.value_counts()

requisicao_info = dados_requisicao_item.groupby('ID_REQUISICAO')['DS_STATUS_ITEM'].value_counts().unstack().fillna(0)
requisicao_info = requisicao_info[requisicao_info['Liberado pelo sistema'] == 0]
requisicao_info.drop(columns=['Liberado pelo sistema'], inplace=True)
requisicao_info
```

```python
quantidade_registros = pd.DataFrame(dados_requisicao_item.groupby('ID_REQUISICAO').size(),columns=['Count'])
quantidade_registros
```

```python
;
# 20 requisicoes em desacordo com criterio tecnico
# 50 liberados pelo usuario
# 50 nao liberado

requisicao_info = requisicao_info.merge(quantidade_registros, how='left', left_index=True, right_index=True)


requisicao_info
```

```python

requisicao_info = requisicao_info[requisicao_info['Em anï¿½lise'] == 0]

itens_liberados  = requisicao_info[ requisicao_info['Liberado pelo usuï¿½rio'] == requisicao_info['Count']]
itens_rejeitados = requisicao_info[ requisicao_info['Nï¿½o liberado'] == requisicao_info['Count']]
itens_desacordo  = requisicao_info[ requisicao_info['Em desacordo com os critï¿½rios tï¿½cnicos'] == requisicao_info['Count']]
itens_liberados_rejeitados = requisicao_info[ (requisicao_info['Liberado pelo usuï¿½rio'] != requisicao_info['Count']) & (requisicao_info['Liberado pelo usuï¿½rio'] > 0)]

itens_liberados.shape,itens_rejeitados.shape,itens_desacordo.shape

import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.bar(['Liberado','Rejeitado','Desacordo','itens_liberados_rejeitados'],[itens_liberados.shape[0],itens_rejeitados.shape[0],itens_desacordo.shape[0],
                                                                           itens_liberados_rejeitados.shape[0]])
plt.show()
```

```python
# sort
itens_liberados = itens_liberados.sample(frac=1,random_state=0)
itens_rejeitados = itens_rejeitados.sample(frac=1,random_state=0)
itens_desacordo = itens_desacordo.sample(frac=1,random_state=0)
itens_liberados_rejeitados = itens_liberados_rejeitados.sample(frac=1,random_state=0)

itens_liberados_rejeitados
```

```python

max_itens = 25
itens = [   (itens_liberados, 'Liberado',max_itens) ,
            (itens_rejeitados, 'Rejeitado',max_itens),
            (itens_desacordo,  'Desacordo',max_itens),
            (itens_liberados_rejeitados, 'Liberado_e_ou_Rejeitado',max_itens),
        ]



x , y = [] , []
for item,nome,qtd in itens:
    x +=  item.index[:qtd].values.tolist()
    y += [nome]*qtd

df_sample = pd.DataFrame({'ID_REQUISICAO':x,'STATUS':y})
df_sample.to_csv(f'sample_agosto_{max_itens*4}.csv',index=False)
```
