```python
import pandas as pd
from gensim.models import Word2Vec
import numpy as np

# Passo 1: Pré-processamento dos dados
# Tokenize os itens (separando palavras)
df_merged_all_drop['DS_ITEM_TOKENIZED'] = df_merged_all_drop['DS_ITEM'].astype(str).str.lower().str.split()

# Passo 2: Treinamento do modelo Word2Vec
# Treine o modelo usando os tokens
model = Word2Vec(sentences=df_merged_all_drop['DS_ITEM_TOKENIZED'], vector_size=100, window=5, min_count=1, sg=1, epochs=10)

# Passo 3: Função para obter o embedding médio de cada item
def get_embedding(item_tokens):
    # Obter o vetor médio das palavras, ignorando palavras não presentes no vocabulário
    vectors = [model.wv[word] for word in item_tokens if word in model.wv]
    if len(vectors) == 0:
        return np.zeros(model.vector_size)
    return np.mean(vectors, axis=0)

# Passo 4: Adicionar os embeddings ao DataFrame
df_merged_all_drop['DS_ITEM_EMBEDDING'] = df_merged_all_drop['DS_ITEM_TOKENIZED'].apply(get_embedding)

# Passo 5: Expandir o embedding em múltiplas colunas (opcional)
embedding_df = pd.DataFrame(df_merged_all_drop['DS_ITEM_EMBEDDING'].tolist(), index=df_merged_all_drop.index)
embedding_df.columns = [f'EMBEDDING_{i}' for i in range(embedding_df.shape[1])]

# Passo 6: Criar o novo DataFrame com o nome df_merged_all_drop_embedding
df_merged_all_drop_embedding = pd.concat([df_merged_all_drop, embedding_df], axis=1)

# Passo 7: Remover as colunas temporárias
df_merged_all_drop_embedding.drop(columns=['DS_ITEM_TOKENIZED', 'DS_ITEM_EMBEDDING'], inplace=True)

# Exibir uma amostra do novo DataFrame
df_merged_all_drop_embedding.head()


# Exibir o DataFrame resultante

```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

# Colunas que deseja aplicar o One-Hot Encoding
colunas_para_codificar = ['DS_TIPO_GUIA', 'DS_CARATER_ATENDIMENTO', 'DS_TIPO_INTERNACAO', 'DS_REGIME_INTERNACAO', 'DS_TIPO_SADT', 'DS_TIPO_ITEM', 'DS_TIPO_CONSULTA', 'DS_TIPO_ACOMODACAO']

# Inicialize o OneHotEncoder
encoder = OneHotEncoder(sparse_output=False, drop='first')  # Usando drop='first' para evitar dummy trap (opcional)

# Aplique o encoder nas colunas selecionadas
encoded_array = encoder.fit_transform(df_merged_all_drop_embedding[colunas_para_codificar])

# Crie um DataFrame com os nomes das colunas codificadas
encoded_df = pd.DataFrame(
    encoded_array,
    columns=encoder.get_feature_names_out(colunas_para_codificar),
    index=df_merged_all_drop_embedding.index
)

# Concatene o DataFrame original com o DataFrame codificado
df_onehot_encoder = pd.concat([df_merged_all_drop_embedding, encoded_df], axis=1)

# Remova as colunas originais
df_onehot_encoder.drop(columns=colunas_para_codificar, inplace=True)

# Exiba o resultado
df_onehot_encoder.head()

```

```python
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
from catboost import CatBoostClassifier
import mlflow.catboost


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=5)
```

```python
pca = PCA(n_components=0.95)  # Retendo 95% da variância
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)
```

```python
dataset_name = 'df_onehot_encoder_drop'

# Começando o MLflow para monitorar o experimento
with mlflow.start_run():
    # Definindo sample weights (exemplo: utilizando o inverso da frequência da classe)
    from sklearn.utils.class_weight import compute_sample_weight

    # Calculando sample weights com base na distribuição da variável target
    sample_weights = compute_sample_weight(class_weight='balanced', y=y_train)

    # Treinando o modelo XGBoost com sample weights
    model = XGBClassifier(random_state=42)
    model.fit(X_train_pca, y_train, sample_weight=sample_weights)

    # Fazendo previsões
    y_pred = model.predict(X_test_pca)

    # Calculando a acurácia
    accuracy = accuracy_score(y_test, y_pred)
    accuracy_rounded = round(accuracy, 2)

    # Registrando a acurácia como métrica
    mlflow.log_param('pca_n_components', 0.95)
    mlflow.log_metric('accuracy', accuracy_rounded)
    mlflow.xgboost.log_model(model, 'xgboost_model')
    mlflow.log_param('dataset', dataset_name)

    print(f"Accuracy: {accuracy:.2f}")
```
