```python
import pandas as pd
```

```python
import os
path = "Ginecologia e obstetrícia\\outputs\\"

dfs = [ pd.read_csv('Manual de instruções.csv'),
        pd.read_csv('Tabela-CBHPM-2018.csv'),
      ]

for f in os.listdir(path):
    dfs.append( pd.read_csv(f'{path}/{f}') ) 


len(dfs)

```

```python
def make_metada(row):
    d = pd.DataFrame(row).T
    columns = [i for i in d.columns if i not in ['conteudo','metadata']]
    item = {}
    for c in columns:
        item[c] = d[c].values[0]
    return item

for i in dfs:
    i.fillna('', inplace=True)
    i['metadata'] = i.apply(make_metada, axis=1)
    i.drop(columns=[i for i in i.columns if i not in ['nivel_1', 'nivel_2', 'conteudo','metadata']], inplace=True)

df = pd.concat(dfs, ignore_index=True)
df
```

```python
df['conteudo'] = df['nivel_1'] + ' - ' + df['nivel_2'] + '\n\n' + df['conteudo']
```

```python
df.drop(columns=['nivel_1', 'nivel_2'], inplace=True)
```

```python
df.to_csv('merged_09_10_2024.csv', index=False, encoding='utf-8')
```
